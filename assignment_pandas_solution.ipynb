### Part 3

There are columns called 'gross' and 'budget' representing the gross income and the budget for each movie.

Create a new column called 'profit' which contains the gross profit (i.e. gross - budget) for each movie.

```python
df['profit'] = df['gross']-df['budget']
```

### Part 4

Find the movies with the 10 highest budgets. Print the Name of the movie, along with the budget, gross and profit to the screen for each of these movies.

```python
sorted_df = df.sort_values(by='budget', ascending=False)
sorted_df[['movie_title', 'budget', 'gross', 'profit']].head(10)
```

### Part 6

Check out the the groupBy and agg functions. (These were not covered in class)
pandas.DataFrame.groupby
pandas.core.groupby.DataFrameGroupBy.agg

Use these to get the average budget for each director.
Then display the directors with the 5 largest budgets.


```python
director_df = df.groupby('director_name')
```

```python
agg_df = director_df['budget'].agg(np.mean)
```

```python
agg_df.sort_values(ascending=False).head()
```
